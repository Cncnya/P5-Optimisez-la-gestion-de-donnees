```python
# On calcul les bornes inférieures et supérieurs pour déterminer les outliers

def quartile(v):
  q1 = np.quantile(v, 0.25)
  q3 = np.quantile(v, 0.75)
  
  EIQ = q3 - q1
  LI = q1 - (EIQ*1.5)
  LS = q3 + (EIQ*1.5)

  print("L'écart interquartile est " +str(EIQ))
  print()
  print("La borne supérieure est " +str(LS))
  print()
  print("La borne inférieure est " +str(LI))
  return EIQ, LI, LS
  
```

```python
quartiles = quartile(erp["price"])
```

```python
def outlier(v):
  q1 = np.quantile(v, 0.25)
  q3 = np.quantile(v, 0.75)
  
  EIQ = q3 - q1
  LI = q1 - (EIQ*1.5)
  LS = q3 + (EIQ*1.5)

  i = list(v.index[(v < LI) | (v > LS)])
  val = list(v[i])


  return i , val
```

```python
outliers = outlier(erp["price"])
```

```python
print("Il y'a " + str(np.array(outliers).shape[1]) + " outliers pour les prix.")
```

```python
# Les oultiers concerncent les produits suivant

erp_outliers = erp.loc[(erp['price'] > 83.1) | (erp['price']<-26.49)  ,:]
erp_outliers.sort_values(by='price')
```

```python
# Représentation graphique des outliers

plt.figure(figsize = (16,10))
plt.title("Représentation des Outliers pour les prix des produits selon l'identifiant produit'", fontsize=16)
graph_outliers = sns.scatterplot(data=erp , x= 'product_id', y='price', hue = erp["price"]> 83.1, legend=["brief"])

handles, labels  =  graph_outliers.get_legend_handles_labels()
graph_outliers.legend(handles, ["Prix 'normaux'",'Prix outliers'], loc='upper right')
```
